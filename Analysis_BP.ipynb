# Kaya equation

$$CO_2 = \frac{CO_2}{MWh}\times MWh$$

```python
tepTOkWh = 11630.


dfCO2 = dfDict['Carbon Dioxide Emissions'] # MtCO2
dfCO2.head(20)

World = 109
OECD = 110
NonOECD = 111
Europe = 54
France = 30
Germany = 31
which = World

dates = dfCO2.iloc[1][1:-3]
CO2_Tot = dfCO2.iloc[which][1:-3]*1e6 # result in tCO2
#print(dates)
#print(TotWorld.values[1:-3])
fig, axes = plt.subplots(1, 3, figsize=(25,5))
axes[0].scatter(dates, CO2_Tot)
axes[0].set_ylim(0,CO2_Tot.max()*1.1)
axes[0].grid()
axes[0].set_ylabel('tCO2')
axes[0].set_xlabel('Year')

dfPrimEnergy = dfDict['Primary Energy Consumption'] # Mtep
PrimEnergy_Tot = dfPrimEnergy.iloc[which][1:-3]*1e6*tepTOkWh/1e3 # result is in MWh
#dfPrimEnergy.head(10)
#dfPrimEnergy.tail(14)

ratio = CO2_Tot/PrimEnergy_Tot

axes[1].scatter(dates, ratio)
axes[1].set_ylim(0, ratio.max()*1.1)
axes[1].grid()
axes[1].set_ylabel('tCO2/MWh')
axes[1].set_xlabel('Year')

axes[2].scatter(dates, PrimEnergy_Tot)
axes[2].set_ylim(0,PrimEnergy_Tot.max()*1.1)
axes[2].grid()
axes[2].set_ylabel('MWh')
axes[2].set_xlabel('Year')

plt.figure()
plt.scatter(PrimEnergy_Tot, CO2_Tot)
plt.ylabel("tCO2")
plt.xlabel("MWh")
plt.grid()

print("tCO2 (from 1965 to 2017): %0.2f" % (CO2_Tot[-1]/CO2_Tot[0]))
print("tCO2/MWh (from 1965 to 2017): %0.2f" % (ratio[-1]/ratio[0]))
print("MWh (from 1965 to 2017): %0.2f" % (PrimEnergy_Tot[-1]/PrimEnergy_Tot[0]))
```

```python

mydict = {'World': 109, 'OECD': 110, 'NonOECD': 111, 'Europe': 54, 'France': 30, 'Germany': 31}
print(mydict)

mydf = dfCO2.iloc[1][1:-3].to_frame(name="Date")
for which, idx in mydict.items():
    print(which, idx)
    mydf["PrimE_"+which] = dfPrimEnergy.iloc[idx][1:-3]*1e6*tepTOkWh/1e3 # in MWh
    mydf["CO2_"+which] = dfCO2.iloc[idx][1:-3]*1e6 # in tCO2
    mydf["EmissionPower_"+which] = mydf["CO2_"+which]/mydf["PrimE_"+which] # in tCO2/MWh
mydf = mydf.set_index('Date')
mydf


quantities = ['CO2', 'EmissionPower', 'PrimE']
axisTitle = ['tCO2', 'tCO2/MWh', 'MWh']
countries = ['World', 'OECD', 'NonOECD']
toplot = []
for q in quantities:
    temp = []
    for c in countries:
        temp.append(q+'_'+c)
    toplot.append(temp)
    
print(toplot)

fig, axes = plt.subplots(1, 3, figsize=(25,5))
for i, t in enumerate(toplot):
    print(i, t)
    max = 0
    for col in toplot[i]:
        if mydf[col].max() > max:
            max = mydf[col].max()
    mydf.plot(y=t, grid=True, ax=axes[i])
    axes[i].set_ylabel(axisTitle[i])
    axes[i].set_ylim(0, max*1.15)
```
